```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
df= pd.read_csv('/content/blood_pressure.csv')
df[['bp_before','bp_after']].describe()
df.head(5)
```

```python
ttest,pval = stats.ttest_rel(df['bp_before'], df['bp_after'])
print(pval)
```

```python
if pval<0.05:
  print('reject null hypothesis')
else:
  print('accept null hypothesis')


```

```python
df_chi = pd.read_csv('/content/BCSprep.csv')
df_chi.head()
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df_chi['lnpos_YN'], df_chi['status'])
```

```python
print("Contingency Table with Frequencies:")
print(contingency_table)
```

```python
Observed_Values = contingency_table.values
print("Observed Values :-\n",Observed_Values)
```

```python
b=stats.chi2_contingency(contingency_table)
print(b,'\n')
Expected_Values = b[3]
print('Expected Values :-\n',Expected_Values)
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
print('chi-square statistic:-',chi_square_statistic)
```

```python
no_of_rows=len(contingency_table.iloc[0:2,0])
no_of_columns=len(contingency_table.iloc[0,0:2])
df11=(no_of_rows-1)*(no_of_columns-1)
print('Degree of Freedom:-',df11)
alpha = 0.05
```

```python
critical_value=chi2.ppf(q=1-alpha,df=df11)
print('critical_value:',critical_value)

```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=df11)
print('p-value:',p_value)
```

```python
print('Significance level: ',alpha)
print('Degree of Freedom: ',df11)
print('chi-square statistic: ',chi_square_statistic)
print('critical_value:',critical_value)
print('p-value:',p_value)
```

```python

if chi_square_statistic>=critical_value:
  print('Reject H0,There is a relationship between 2 categorical variables')
else:
  print('Retain H0,There is no relationship between 2 categorical variables')

if p_value<=alpha:
  print('Reject H0,There is a relationship between 2 categorical variables')
else:
  print('Retain H0,There is no relationship between 2 categorical variables')

```

```python


import numpy as np

# Extract data from DataFrame
x = df['bp_before'].values
y = df['bp_after'].values

# Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate deviations from the mean
x_dev = x - x_mean
y_dev = y - y_mean

# Calculate the covariance
covariance = np.sum(x_dev * y_dev)

# Calculate the variance of x
x_variance = np.sum(x_dev ** 2)

# Calculate the slope of the regression line
slope = covariance / x_variance

# Calculate the intercept of the regression line
intercept = y_mean - slope * x_mean

# Print the results
print("Regression analysis results:")
print("Slope:", slope)
print("Intercept:", intercept)
```
