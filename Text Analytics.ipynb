```python
# a) Perform text preprocessing techniques (tokenization, stemming, stop word removal) on textual data.
```

```python
# Importing the required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
text_data = """Nelson Mandela and Martin Luther King Jr. both fought for racial equality. Although Luther King
was an American citizen and Mandela a native South African, their dreams were the same. The intolerance of white
people towards black co-inhabitants was the catalyst for years of activism. Their battles were tumultuous -
Nelson Mandela was arrested in 1962 for treason. Consequently, he was incarcerated for twenty-seven years.
In contrast, although Luther King was detained several times, the longest jail sentence he received was
four months. Both Mandela and Luther King have been awarded the Nobel Peace Prize for their dedication to
improving civil rights for black people. Martin Luther King Jr. led many demonstrations against racism.
He delivered his message in a non-violent manner, although other members of his movement later engaged in
less peaceful protests. In contrast, Mandela initially avoided violence, but resorted to it following the
government's massacre of unarmed black Africans. Martin Luther King's famous 1963 speech, "I Have a Dream",
inspired many African-Americans to envisage a better future. He was assassinated in 1968. During Nelson Mandela's
best known speech he recited “Our Deepest Fear", an inspirational poem written by Marianne Williamson.
In 1994, Mandela became the first black president of South Africa; he was the first president elected by the
people."""
```

```python
# Tokenization
import nltk
nltk.download('punkt')

tokens = word_tokenize(text_data)
print("Tokens:", tokens)
print("Number of tokens:", len(tokens))
```

```python
# Punchuation removal
tokens = [word for word in tokens if word.isalnum()]
print("Tokens after removing punctuation\n", tokens)
print("Number of tokens:", len(tokens))
```

```python
# Stop word removal
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

print("Stop Words in English:", stop_words)
print()

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("Filtered Tokens:", filtered_tokens)
print("Number of tokens:", len(filtered_tokens))
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)
```

```python
# Lemmatization
import nltk
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized Tokens:", lemmatized_tokens)
```

```python
# b) Perform Sentiment Analysis on IMDB Dataset with two fields: review and sentiment.
```

```python
# Load the IMDB dataset
import re
import pandas as pd
imdb_data = pd.read_csv("/content/IMDB Dataset.csv")

# Dropping 20000 rows to reduce the size of the dataset
imdb_data = imdb_data.drop(imdb_data.index[20000:])
print(imdb_data.head())
```

```python
imdb_data.info()
```

```python
# Rename the columns
imdb_data.columns = ['review', 'sentiment']
print(imdb_data.head())
```

```python
# Null value check
print(imdb_data.isnull().sum())
```

```python
imdb_data.nunique()
```

```python
# Function to preprocess the review text
def preprocess_review_text(review):
    # Remove HTML tags
    review = re.sub('<.*?>', '', review)
    tokens = word_tokenize(review)  # Tokenization
    tokens = [word for word in tokens if word.isalnum()]    # Punchuation removal
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]   # Stop word removal
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]   # Lemmatization
    cleaned_review = ' '.join(lemmatized_tokens)
    return cleaned_review
```

```python
# Preprocess the review text
# This takes time to process 40000 records
imdb_data['cleaned_review'] = imdb_data['review'].apply(preprocess_review_text)
```

```python
imdb_data.head()
```

```python
# Drop the review column
imdb_data.drop('review', axis=1, inplace=True)
```

```python
# Word Cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stop_words, min_font_size = 10).generate(' '.join(imdb_data['cleaned_review']))
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
# Count plot
import seaborn as sns
sns.countplot(x='sentiment', data=imdb_data)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# tf-idf vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

x = imdb_data['cleaned_review']
y = imdb_data['sentiment']

vector = TfidfVectorizer()
x = vector.fit_transform(x)
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Applying the Logistic Regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)

print("Accuracy of Logistic Regression model:", accuracy_score(y_test, y_pred))
```

```python
# Confusion Matrix heat map
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
